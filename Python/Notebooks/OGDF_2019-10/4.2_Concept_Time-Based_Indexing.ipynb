# From a string, make it a "Datetime" format.

There is a special format called DateTime. If we use that, we can do a lot of time-related calculations easily.

```python
cc2_file = 'data/LNG_project/CC2_1_year_dataset_2017-2018.csv'
cc2 = pd.read_csv(cc2_file)
cc2
```

```python
#cc2.dtypes
#convert to a nice datetime format...otherwise it is just a string
timecol = 'Time' # 'Time' is the name of the column in our df. We are storing it in a variable called timecol

cc2[timecol] = pd.to_datetime(cc2[timecol]) # converting it
#cc2[timecol]
cc2[timecol].dtype
```

```python
cc2[timecol].head()
cc2['Time'].dtype #ns is nanoseconds
```

# Convert the data frame to have Time Index

```python
cc2.columns
cc2.set_index('Time', inplace=True)
cc2
```

# Time Series based analysis

```python
cc2['2017-04':'2017-07'].tail(3)
```

```python
cc2 = cc2.applymap(float) # make the columns from OBJECT to FLOAT
cc2.dtypes
```
